```python
url = 'https://data.lacity.org/api/views/2nrs-mtv8/files/4591b6bf-5846-4de0-9fb0-8780a77a036c?download=true&filename=MO_CODES_Numerical_20191119.pdf'
response = requests.get(url)
response.status_code
```

```python
# Translating the Mocodes from their code to their description according to LAPD, 
# The descroption is in PDF file so the information is extracted according to the files structure 
mocodes = {}
for page in pdfplumber.open(io.BytesIO(response.content)).pages:
    pg = page.extract_text()
    pg = pg.split('\n')[3:-1]
    for code in pg:
        mocodes[code[:4]] = code[5:]
crime['Mocodes'] = crime['Mocodes'].fillna('').str.split()

def process_mo_row(code_list):
    if not isinstance(code_list, list):
        return "Unknown"
    if len(code_list) == 0:
        return "Unknown"
    descriptions = [mocodes.get(str(code), "Code Not Found") for code in code_list]
    return ", ".join(descriptions)


crime['Mocode Text'] = crime['Mocodes'].apply(process_mo_row)
```

```python
# Check that one or more columns in Crm Cd 1 to 4 can be filled simultaneously
severity_cols = ['Crm Cd 1', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4']
crime_counts = crime[severity_cols].notna().sum(axis=1)
multi_crime_mask = crime_counts > 1
overlap_df = crime[multi_crime_mask]
print(f"Total rows with more than one Crime Code: {len(overlap_df)}")
if not overlap_df.empty:
    display(overlap_df[severity_cols].info())
```

```python
# We will create a column to rank the severity of the crimes based on the columns description and structure logic
#1) map the part 1-2 to flip them (2 is more severe)

#2) for every column Crm Cd 1 to 4 if the value is not null i will replace with prime numbers that are linearly independent, In this case I chose, Crm 1 :11, crm 2: 7, crm 3: 3, crm 4: 2

#3) sum the 4 columns together (if there is a combination it will have a higher score) and multiply by the mapped (part 1-2) column
```

```python
crm1 = np.where(crime['Crm Cd 1'].isnull(),0,11)
crm2 = np.where(crime['Crm Cd 2'].isnull(),0,7)
crm3 = np.where(crime['Crm Cd 3'].isnull(),0,3)
crm4 = np.where(crime['Crm Cd 4'].isnull(),0,2)
part21 = crime['Part 1-2'].map({1: 2, 2: 1}).fillna(1)
sev_score = part21*(crm1 + crm2 + crm3 + crm4)

unique_sev_scores = sorted(sev_score.unique())
for i in range(len(unique_sev_scores)):
    sev_score.replace(unique_sev_scores[i],i+1,inplace=True)

crime['Severity Score'] = sev_score
```

```python
# Change the date reported and date occured to the format DD/MM/YYYY
crime['Date Rptd'] = pd.to_datetime(crime['Date Rptd'])
crime['Date_Reported'] = crime['Date Rptd'].dt.strftime('%d/%m/%Y')

crime['DATE OCC'] = pd.to_datetime(crime['DATE OCC'])
crime['Date_Occured'] = crime['DATE OCC'].dt.strftime('%d/%m/%Y')
```

```python
# Fix the Time of Occurrance column
crime['TIME OCC'] = crime['TIME OCC'].astype(str).str.zfill(4)
```

```python
# Map the victim descent column to their corresponding description
ethnicity = {'A': 'Other Asian', 'B':'Black', 'C':'Chinese','D':'Cambodian','F':'Filipino', 'G':'Guamanian',
             'H':'Hispanic/Latin/Mexican','I':'American Indian/Alaskan Native','J':'Japanese' ,'K':'Korean','L':'Laotian',
             'O':'Other', 'P':'Pacific Islander','S':'Samoan', 'U':'Hawaiian','V':'Vietnamese','W':'White','X':'Unknown','Z':'Asian Indian'}
crime['Vict Descent'] = crime['Vict Descent'].map(ethnicity)
```

```python
# Rearrange the columns while getting rid of the ones who dont provide information we can analyze
crime = crime[['DR_NO','Date_Occured','TIME OCC','Date_Reported','Crm Cd Desc','Weapon Desc','Mocode Text','Severity Score','Vict Sex','Vict Age',
      'Vict Descent','AREA NAME', 'LOCATION', 'Cross Street','Premis Desc','LAT','LON']]
crime
```

```python
# Rename the columns
crime.rename(columns={'DR_NO':'File ID',
                      'Date_Occured':'Date Occured',
                      'TIME OCC':'Time Occured',
                      'Date_Reported':'Date Reported',
                      'Crm Cd Desc':'Crime Description',
                      'Weapon Desc':'Weapon',
                      'Vict Sex':'Victim Sex',
                      'Vict Age':'Victim Age',
                      'Vict Descent':'Victim Descent',
                      'AREA NAME':'Area',
                      'LOCATION':'Location',
                      'Premis Desc':'Premis',
                      'LAT':'Latitude',
                      'LON':'Longitude'},
                       inplace=True)
```
